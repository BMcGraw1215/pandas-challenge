```python
#Written report at end of file.
import os
import csv
import pandas as pd
```

```python
school_data = "Resources/schools_complete.csv"
student_data = "Resources/students_complete.csv"
```

```python
school_data = pd.read_csv(school_data)
student_data = pd.read_csv(student_data)
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])    
```

```python
school_data_complete
```

```python
#Total Number of Schools
TotalNumSch= school_data["school_name"].size
```

```python
#Total Number of Students
TotalNumStu = school_data_complete["Student ID"].size
```

```python
#Total Budget
TotalBudget = school_data["budget"].sum()

TotalBudgetF = '${:,}'.format(TotalBudget)
```

```python
#Returns the average math score for all students
AvgMathSco = school_data_complete["math_score"].mean()

AvgMathScoS = round(AvgMathSco, 4)
```

```python
#Returns the average reading score for all students
AvgReadSco = school_data_complete["reading_score"].mean()

AvgReadScoS = round(AvgReadSco, 4)
```

```python
#Return the Percent Value of students who pass Maths
MathPassingTF = school_data_complete["math_score"] >= 70     #returns list of T/F (If score is greater than 70, true)

MathPassing = 0
for _ in range(0, len(MathPassingTF)):                       #goes through T/F list. If True, MathPassing + 1. 
    if(MathPassingTF[_] == True):                            #This is the total number of students who passed math.
        MathPassing = MathPassing + 1
    
MathPassingPctg = (MathPassing / TotalNumStu) * 100          #converts to percentage.

MathPassingPctgR = round(MathPassingPctg, 4)
```

```python
#Return the Percent Value of students who pass Reading
ReadPassingTF = school_data_complete["reading_score"] >= 70     #returns list of T/F (If score is greater than 70, true)

ReadPassing = 0
for _ in range(0, len(ReadPassingTF)):                       #goes through T/F list. If True, ReadPassing + 1. 
    if(ReadPassingTF[_] == True):                            #This is the total number of students who passed reading.
        ReadPassing = ReadPassing + 1
    
ReadPassingPctg = (ReadPassing / TotalNumStu) * 100          #converts to percentage.

ReadPassingPctgR = round(ReadPassingPctg, 4)
```

```python
#Return the Percent Value of students who passed
TotalPassing = 0
for _ in range(0, len(ReadPassingTF)):                       #goes through BOTH T/F list. If Both are True, TotalPassing + 1. 
    if(ReadPassingTF[_] == True and MathPassingTF[_] == True):         
        TotalPassing = TotalPassing + 1
        
TotalPassingPctg = (TotalPassing / TotalNumStu) * 100

TotalPassingPctgR = round(TotalPassingPctg, 4)
```

```python
#Dictionary and DataFrame 
DataFrameDict = {'Total Schools' : TotalNumSch, 'Total Students' : TotalNumStu, 'Total Budget' : TotalBudgetF , 'Average Math Score' : AvgMathScoS, 'Average Reading Score' : AvgReadScoS , '% Passing Math' : MathPassingPctgR, '% Passing Reading' : ReadPassingPctgR, '%Overall Passing' : TotalPassingPctgR}

District_df = pd.DataFrame(DataFrameDict, index = [1])

District_df
```

```python
#Makes a list of unique schools
SchNames = school_data_complete["school_name"].unique()
Setlist = set(school_data_complete["school_name"].unique())
```

```python
#Declares Variables, runs a for statement to pull all the data from all schools to create a summary
gradelist = ['9th', '10th', '11th', '12th']
    
ninethm = []
tenthm = []
eleventhm = []
twelvethm = []


ninethr = []
tenthr = []
eleventhr = []
twelvethr = []

y = 0
b = []   #type
sz = []  #size
sb = []  #student budget
bu = []  #budget

ma = []  #math average
ra = []  #reading average
mp = []
rp = []
tp = []
Mathverage = []
Readverage = []

#This method is highly inefficient... will likely review. 
#Missing 1-2 concepts or formulas to use DFs/Series' effectively, cannot make simple manipulations work...
#resorted to appending lists per DF, manipulating those, and turning them into Series'
for x in SchNames:
    
    school_df = school_data_complete.set_index('school_name')
    indschool_df = school_df.loc[x]                   #for every for loop, indschool_df is a new school
    n = school_data_complete["school_name"].unique()
    
    if(any(indschool_df['type'] == "District")):      #Checks District
        b.append('District')
    else:
        b.append("Charter")
    sz.append(indschool_df['size'].mean())                   ####
    bu.append(indschool_df['budget'].mean())                 #updates lists to add school characteristics    
    sb.append(bu[y]/sz[y])                                   #
                                                             #
    ma.append(indschool_df['math_score'].mean())             #
    ra.append(indschool_df['reading_score'].mean())          #
    
    IndMathPassing = indschool_df["math_score"] >= 70            #returns list of T/F (If score is greater than 70, true)

    MathPassingi = 0
    for _ in range(0, len(IndMathPassing)):                       #goes through T/F list. If True, MathPassing + 1. 
        if(IndMathPassing[_] == True):                            #This is the total number of students who passed math.
            MathPassingi = MathPassingi + 1
    mp.append(round((MathPassingi/sz[y]*100), 4))
    
    IndReadingPassing = indschool_df["reading_score"] >= 70   
    
    ReadingPassingi = 0
    for _ in range(0, len(IndReadingPassing)):                       #goes through T/F list. If True, ReadingPassing + 1. 
        if(IndReadingPassing[_] == True):                            #This is the total number of students who passed math.
            ReadingPassingi = ReadingPassingi + 1
    rp.append(round((ReadingPassingi/sz[y]*100), 4))
    
    TotalPassingi = 0
    for _ in range(0, len(IndReadingPassing)):                       #goes through T/F list. If True, TotalPassing + 1. 
        if(IndReadingPassing[_] == True and IndMathPassing[_] == True):                            #This is the total number of students who passed math.
            TotalPassingi = TotalPassingi + 1
    tp.append(round((TotalPassingi/sz[y]*100), 4))
    
    
    for x2 in gradelist:                
        Gradescore_df = indschool_df[['grade', 'math_score']]      #Makes DF for grades and scores per school
        Gradescore_df = Gradescore_df.set_index('grade')      #new DF for specific data
        MathScore_df = Gradescore_df.loc[x2]                  #pulls per school_name
        Mathverage.append(MathScore_df['math_score'].mean())         #appends the average for that school
    
    for x2 in gradelist:
        Gradescore_df = indschool_df[['grade', 'reading_score']]     
        Gradescore_df = Gradescore_df.set_index('grade')
        ReadScore_df = Gradescore_df.loc[x2]                 
        Readverage.append(ReadScore_df['reading_score'].mean())
    
    
    
    ninethm.append(Mathverage[0])         #Appends the Math and Reading scores per grade in a way that can be used
    tenthm.append(Mathverage[1])          #for Series/Dataframes
    eleventhm.append(Mathverage[2])       #
    twelvethm.append(Mathverage[3])       #
    
    ninethr.append(Readverage[0])         #
    tenthr.append(Readverage[1])          #
    eleventhr.append(Readverage[2])       #
    twelvethr.append(Readverage[3])       #
    
    
    y = y + 1                    #Resets certain variables for loop
    Mathverage.clear()           #
    Readverage.clear()           #
    
    
#turns all lists into Series' and formats them.

n = pd.Series(n)
b = pd.Series(b)
bu = pd.Series(bu).map('${:,.2f}'.format)
sz = pd.Series(sz).round(0)
sb = pd.Series(sb).map('${:,.2f}'.format)
ma = pd.Series(ma).round(4)
ra = pd.Series(ra).round(4)
mp = pd.Series(mp).round(4)
tp = pd.Series(tp).round(4)
ninethm = pd.Series(ninethm).round(4)
tenthm = pd.Series(tenthm).round(4)
eleventhm = pd.Series(eleventhm).round(4)
twelvethm = pd.Series(twelvethm).round(4)
ninethr = pd.Series(ninethr).round(4)
tenthr = pd.Series(tenthr).round(4)
eleventhr = pd.Series(eleventhr).round(4)
twelvethr = pd.Series(twelvethr).round(4)
```
